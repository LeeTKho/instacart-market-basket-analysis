```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

plt.style.use('ggplot')
```

```python
# import all data
aisles = pd.read_csv("../data/aisles.csv")
departments = pd.read_csv("../data/departments.csv")
op_prior = pd.read_csv("../data/order_products__prior.csv")
op_train = pd.read_csv("../data/order_products__train.csv")
orders = pd.read_csv("../data/orders.csv")
products = pd.read_csv("../data/products.csv")
```

## **CUSTOMER SEGMENTATION**

(Adapted from https://github.com/jrkreiger/instacart-analysis/blob/master/instacart-analysis.ipynb)

**PART 1: FEATURE ENGINEERING**

In order to perform RFM segmentation, we have to perform some feature engineering to obtain proxy features for each factor.

"RFM" = "Recency, frequency, monetary value"

RFM feature engineering:
1. Recency - Number of orders (n_orders)
2. Frequency - Average days between orders (avg_days)
3. Monetary value - Average number of items ordered per order (avg_num_prods)

**PART 2: K-MEANS CLUSTERING**

After performing the feature engineering and data preprocessing, we will use k-means clustering to determing clusters.

We will use the elbow technique as well as t-SNE plots to find the ideal k value for our clusters.

**PART 3: INTERPRET SEGMENTS & AND USE SEGMENTS FOR RECOMMENDER SYSTEM**

Finally, we will analyze and interpret the final clusters and use these clusters to build our recommender systems.

**FEATURE ENGINEERING:**

```python
# merge dataframes as necessary
products_aisle = products.merge(aisles[['aisle_id','aisle']],on=['aisle_id','aisle_id'])
products_aisle_dept = products_aisle.merge(departments[['department_id','department']],on=['department_id','department_id'])
```

```python
# orders = orders.drop('eval_set',axis=1)

cust_orders = op_prior.merge(
                    orders[['order_id','user_id','order_number','order_dow','order_hour_of_day','days_since_prior_order']],
                    on=['order_id','order_id'],suffixes=('',''))

cust_orders = cust_orders.merge(products_aisle_dept[['product_id','aisle_id','department_id']],on=['product_id','product_id'])
```

```python
# create cust_profile dataframe for our engineered features
cust_profile = cust_orders.sort_values('user_id')
```

```python
cust_profile.head()
```

```python
# calculate number of orders for each user
cust_profile = cust_profile.groupby('user_id')[['order_id']].nunique().reset_index()
cust_profile.columns = ['user_id','n_orders']
```

```python
# calculate average days between orders for each user
cust_profile['avg_days'] = orders.groupby('user_id')['days_since_prior_order'].mean().reset_index()['days_since_prior_order']
```

```python
# calculate average order size for each user
cust_profile['avg_num_prods'] = np.array(cust_orders.groupby(['user_id','order_number']).nunique().groupby('user_id').mean()['product_id'])
```

```python
cust_profile.head()
```

```python
cust_profile.user_id.nunique()
```

**SCALE & TRANSFORM FEATURES FOR CLUSTERING**

```python
# taking random sample of n = 40_000 for clustering; this is necessary due to computational limitations
X_orig = cust_profile.drop(['user_id'],axis=1).sample(40000, random_state=1)
```

```python
# check feature distributions
plt.figure(figsize=(12,6))

plt.subplot(2, 2, 1)
plt.title("Average Number of Products")
plt.hist(X_orig['avg_num_prods'],bins=15)

plt.subplot(2, 2, 2)
plt.title("Number of Orders")
plt.hist(X_orig['n_orders'],bins=15)

plt.subplot(2, 2, 3)
plt.title("Average Days Between Orders")
plt.hist(X_orig['avg_days'],bins=15)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))

plt.subplot(1, 1, 1)
plt.title("Number of Orders (Original)")
plt.hist(X_orig['n_orders'],bins=20)

plt.tight_layout()
plt.show()
```

```python
# log transform n_orders to rectify strong positive skew for clustering
X_log = X_orig.copy()
X_log['n_orders'] = np.log(X_orig['n_orders'])
```

```python
plt.figure(figsize=(6,8))

plt.subplot(2, 1, 1)
plt.title("Number of Orders (Original)")
plt.hist(X_orig['n_orders'],bins=20)
plt.ylabel("# of Users")
plt.xlabel("# of Orders")

plt.subplot(2, 1, 2)
plt.title("Number of Orders (Log-transformed)")
plt.hist(X_log['n_orders'],bins=20)
plt.ylabel("# of Users")
plt.xlabel("Log-transformed # of Orders")

plt.tight_layout()
plt.show()
```

```python
# scale features for clustering
scaler = StandardScaler()
X = scaler.fit_transform(X_log)

X = pd.DataFrame(X,columns = cust_profile.drop(['user_id'],axis=1).columns)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))

visualizer.fit(X)        # fit the data to the visualizer
visualizer.show()        # finalize and render the figure
```

**t-SNE FOR CLUSTER SELECTION:**

```python
# define a function to calculate and plot specified # of clusters
def kmeans(df, clusters_number):
    '''
    Implement k-means clustering on dataset
    
    INPUT:
        dataset : dataframe. Dataset for k-means to fit.
        clusters_number : int. Number of clusters to form.
        end : int. Ending range of kmeans to test.
    OUTPUT:
        Cluster results and t-SNE visualisation of clusters.
    '''   
    kmeans = KMeans(n_clusters = clusters_number, random_state = 1)
    kmeans.fit(df)

    # extract cluster labels
    cluster_labels = kmeans.labels_
        
    # create a cluster label column in original dataset
    df_new = df.assign(cluster = cluster_labels)
    
    # initialise TSNE
    model = TSNE(random_state=1)
    transformed = model.fit_transform(df)
    
    # plot t-SNE
    plt.title('Flattened Graph of {} Clusters'.format(clusters_number))
    sns.scatterplot(x=transformed[:,0], y=transformed[:,1], 
                    hue=cluster_labels, style=cluster_labels, palette="Set1")
    
    return df_new, cluster_labels
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1, 3, 1)
k3 = kmeans(X, 3)

plt.subplot(1, 3, 2)
k5 = kmeans(X, 4)

plt.subplot(1, 3, 3)
k6 = kmeans(X, 5)

plt.tight_layout()
plt.savefig('various_k_clusters.png', format='png', dpi=300, pad_inches=2.0)
plt.show()
```

**INTERPRET CLUSTERS**

```python
# convert clusters to DataFrame with appropriate index and column names
cluster_df = pd.DataFrame(k3[0])
cluster_df['user_id'] =cust_profile.user_id

cluster_df.head()
```

```python
cluster_melt = pd.melt(cluster_df.reset_index(),
                       id_vars=['user_id', 'cluster'],
                       value_vars=['n_orders', 'avg_days', 'avg_num_prods'],
                       var_name='metric',
                       value_name='value')
```

```python
# snake plot to compare clusters
palette = ['b','r','g']
plt.figure(figsize=(6,5))
sns.pointplot(x='metric', y='value', data=cluster_melt, hue='cluster', 
              palette=palette)
plt.xlabel('')
plt.ylabel('Value')
plt.yticks([])
plt.title('Three Customer Segments')
sns.despine()
plt.tight_layout()
plt.savefig('snake_plot.png', dpi=300, pad_inches=2.0)
plt.show();
```

Cluster 0: Small infrequent orders

Cluster 1: Small frequent orders

Cluster 2: Large infrequent orders

**GET CLUSTER LABELS FOR ALL USERS**

```python
kmeans_3 = KMeans(n_clusters = 3, random_state = 1)
kmeans_3.fit(X)
```

```python
X_all = cust_profile.drop(['user_id'],axis=1)

clusters_all = kmeans_3.predict(X_all)
```

```python
clusters_df = pd.DataFrame({'user_id':cust_profile['user_id'],'clusters':clusters_all})
```

```python
cluster_0 = clusters_df[clusters_df.clusters==0]['user_id'].to_list()
cluster_1 = clusters_df[clusters_df.clusters==1]['user_id'].to_list()
cluster_2 = clusters_df[clusters_df.clusters==2]['user_id'].to_list()
```

```python
# pickling the lists of user_ids for each cluster
pd.to_pickle(cluster_0,'../data/cluster_0')
pd.to_pickle(cluster_1,'../data/cluster_1')
pd.to_pickle(cluster_2,'../data/cluster_2')
```
